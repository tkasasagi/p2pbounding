```python
from PIL import Image

directory = 'small_sample/'
ratio = 0.5
```

```python
from os import listdir
filelist = (listdir(directory))
print(len(filelist))
```

```python
for num in range(len(filelist)):
    #Open Image
    img = Image.open(directory + filelist[num])
    print(img.size)
    
    sizex = round(img.size[0] * ratio)
    sizey = round(img.size[1] * ratio)
    
    #print(sizex, sizey)
    
    #Resize image
    resize_img = img.resize((sizex, sizey))
    
    
    filename = directory + filelist[num]
    
    resize_img.save(filename)
    
    #Close Image
    img.close()
    print(num)
```
