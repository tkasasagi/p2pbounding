```python
from PIL import Image

directory = 'small_sample/'
save_directory = 'small_tile/'
```

```python
from os import listdir
filelist = (listdir(directory))
print(len(filelist))
```

```python
for num in range(len(filelist)):
    #print(num)
    #print(filelist[num])
    
    #Open Image
    img = Image.open(directory + filelist[num])
    
    #Cut and Save
    x, y = img.size
    width = int(x/256)
    height = int(y/256)
    
    for j in range(0, height):
        topy = j * 256
        #print("this is y loop" , topy)    
        for i in range(0, width):
            topx = i * 256
            upper = [topx, topy]
            bottom = [(topx + 256), (topy + 256)]
        
            #Insert image and crop and save
            dim = (upper[0], upper[1], bottom[0], bottom[1])
            crop_image = img.crop(dim)
            filename = save_directory + str(num) + str(upper) + str(bottom) + ".jpg"
            print(filename)
            crop_image.save(filename)
        
        #print("Save Image " , upper, bottom)
    #Close Image
    img.close()
```
